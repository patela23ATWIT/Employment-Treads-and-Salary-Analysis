## Finding Significant Gender Based Variations in Pay

```python
# Find the amount of people in each gender group within the dataset
import pandas as pd

data = pd.read_csv('Salary_Data.csv')

# Count for each gender
gender_count = data['Gender'].value_counts()

print(gender_count)
```

```python
# Find median salary for each gender group 
import pandas as pd

data = pd.read_csv('Salary_Data.csv')

# Calculate the median salary for each gender
median_salary_gender = data.groupby('Gender')['Salary'].median()

print(median_salary_gender)
```

```python
# Find average salary by gender group and plot it
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('Salary_Data.csv')

# Calculate the average salary by gender
average_salary_gender = data.groupby('Gender')['Salary'].mean()

print(average_salary_gender)

# Plotting the average salary by gender
average_salary_gender.plot(kind='bar', color=['blue', 'pink', 'red'])
plt.title('Average Salary by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Salary (USD $)')
plt.show()
```

```python
# Find average salary for the Top 50 Males and Females and plot it
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('Salary_Data.csv')

# Sort the data by 'Salary' in descending order to get the top salaries
sorted_data = data.sort_values(by='Salary', ascending=False)

# Get the top 50 salaries
top_50_males = sorted_data[sorted_data['Gender'] == 'Male'].head(50)
top_50_females = sorted_data[sorted_data['Gender'] == 'Female'].head(50)

# Calculate the average salary
average_salary_50_males = top_50_males['Salary'].mean()
average_salary_50_females = top_50_females['Salary'].mean()

# Print the average salary
print(f"Average Salary for Top 50 Males: {average_salary_50_males}")
print(f"Average Salary for Top 50 Females: {average_salary_50_females}")

# Plotting the average salary
plt.bar(['Males', 'Females'], [average_salary_50_males, average_salary_50_females], color=['blue', 'pink'])
plt.title('Average Salary for Top 50 Males and Females')
plt.xlabel('Gender')
plt.ylabel('Average Salary (USD $)')
plt.show()
```

## Find relationships between work experience and salary

```python
# Find average salary based on work experience and plot it
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('Salary_Data.csv')

# Calculate average salary based on experience
average_salary_experience = data.groupby('Years of Experience')['Salary'].mean()

# Create a color pallete of blues that get darker for every incrimentation 
color_blue = [plt.cm.Blues(0.5 + i / (2.0 * len(average_salary_experience))) for i in range(len(average_salary_experience))]
average_salary_experience.plot(kind='bar', figsize=(12, 7), color=color_blue)

plt.title('Average Salary by Years of Work Experience')
plt.xlabel('Years of Work Experience')
plt.ylabel('Average Salary (USD $)')

# Convert x-axis labels to face upright 
plt.xticks(rotation=0)

plt.tight_layout()
plt.show()
```

## Using SciKit Learn create a trend line to draw a relationship between salary and different levels of education

```python
# Create a trend line based on salary and education level using SciKit Learn libraries 
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

data = pd.read_csv('Salary_Data.csv')

# Combine "Master's" and "Master's Degree" into a single category
data['Education Level'] = data['Education Level'].replace("Master's Degree", "Master's")
# Combine "Bachelor's" and "Bachelor's Degree" into a single category
data['Education Level'] = data['Education Level'].replace("Bachelor's Degree", "Bachelor's")

# Manually set the 'Education Level' by difficulty 
education_status = {'High School': 0, "Bachelor's": 1, "Master's": 2, 'PhD': 3}
data['Education Level Encoded'] = data['Education Level'].map(education_status)

x = data[['Education Level Encoded']]
y = data['Salary']

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Create a linear regression model
model = LinearRegression()

# Fit the model
model.fit(X_train, Y_train)

# Make predictions the test
Y_pred = model.predict(X_test)

# Plot the trend line over the sorted data points
plt.scatter(X_test, Y_test, color='black')
plt.plot(X_test, Y_pred, color='blue', linewidth=3)
plt.xlabel('Education Level')
plt.ylabel('Salary (USD $)')
plt.title('Trend Line of Salary by Education Level')

# Print x-axis labels
plt.xticks(ticks=np.arange(len(education_status)), labels=list(education_status.keys()))
plt.show()
```
